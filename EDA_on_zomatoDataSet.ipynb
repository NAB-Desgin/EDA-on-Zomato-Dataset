```python
df=df.drop(['url', 'address','phone','menu_item','dish_liked','reviews_list'], axis=1)
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
df['rate'].unique()
```

```python
def handlerate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value=str(value).split('/')
        value=value[0]
        return float(value)
df['rate'] = df['rate'].apply(handlerate)
df['rate'].head()
```

```python
df
```

```python
df['rate'].isnull().sum()
```

```python
df['rate'].fillna(df['rate'].mean(), inplace=True)
df['rate'].isnull().sum()
```

```python
df.dropna(inplace=True)
df.shape
df.head()
```

```python
df.rename(columns={'approx_cost(for two people)':'cost2plates',	'listed_in(type)':'Type'}, inplace=True)
```

```python
df.drop(['listed_in(city)'], axis=1, inplace=True)
```

```python
df['cost2plates'].unique()
```

```python
def handlecomma(value):
    value=str(value)
    if ',' in value:
        value=value.replace(',','')
        return float(value)
    else:
        return float(value)

df['cost2plates']=df['cost2plates'].apply(handlecomma)
```

```python
df['rest_type'].unique()
```

```python
df['rest_type'].value_counts()
```

```python
rest_types=df['rest_type'].value_counts(ascending = False)
rest_types
```

```python
rest_types_lessthan1000  = rest_types[rest_types<1000]
rest_types_lessthan1000
```

```python
def handleRest(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value

df['rest_type']=df['rest_type'].apply(handleRest)
df['rest_type'].value_counts()
        
```

```python
df.head()
```

```python
df['location'].value_counts()
```

```python
location= df['location'].value_counts(ascending=False)
locLessThen300 = location[location<300]
```

```python
def handleLoc(value):
    if value in locLessThen300:
        return 'others'
    else:
        return value
df['location'] = df['location'].apply(handleLoc) 
df['location'].value_counts()
```

```python
df.head()
```

```python
cuisines=df['cuisines'].value_counts(ascending= False)
cuisinesLessThen100 = cuisines[cuisines<100]

def handlecuisines(value):
    if value in cuisinesLessThen100:
        return 'others'
    else:
        return value

df['cuisines']=df['cuisines'].apply(handlecuisines)
df['cuisines'].value_counts()
        
```
